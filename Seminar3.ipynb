```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptop_price.csv', encoding='windows-1251')
df.head()
```

1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)


```python
df['Cpu'].value_counts()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda row: row.split()[0])
df.sample(50)
```

1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)


```python
df['Memory'].value_counts()
```

```python
# 256GB SSD +  1.0TB Hybrid 
def sum_gb(cell_str):
    gb = 0
    cell_lst = cell_str.split()
    for el in cell_lst:
        if 'GB' in el:
            gb += int(el[:-2])
        elif 'TB' in el:
            gb += float(el[:-2]) * 1024
    return gb


sum_gb('256GB SSD +  1.0TB Hybrid')
            
        
    
```

```python
df['Memory_Amount'] = df['Memory'].apply(sum_gb)
df.sample(10)
```

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)


```python
# 64GB Flash Storage +  1TB HDD
def create_type_mem(cell_str):
    t_m = ''
    cell_lst = cell_str.split()
    for el in cell_lst:
        if 'GB' not in el and 'TB' not in el:
          t_m += el+' '
    return t_m.rstrip()

create_type_mem('64GB Flash Storage +  1TB HDD')
        
    
```

```python
df['Memory_Type'] = df['Memory'].apply(create_type_mem)
df.sample(10)
```

```python
# 64GB Flash Storage +  1TB HDD
def create_type_mem(cell_str):
    t_m = ''
    cell_lst = cell_str.split()
    for el in cell_lst:
        if el.isalpha():
          t_m += el+' '
    return t_m.rstrip()

create_type_mem('64GB Flash Storage +  1TB HDD')
```

1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
df.head()
```
