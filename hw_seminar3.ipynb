```python
avg = df.groupby('bedrooms').agg({'price': 'mean'}).sort_values(by='price', ascending=True)
avg
```

```python
price = df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
price
```

```python
view_waterfront = df.pivot_table(index='view', columns='waterfront', aggfunc='count', values='bedrooms', fill_value = 0)
view_waterfront
```
